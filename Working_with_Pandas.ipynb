```python
#Read the CSV file into a pandas as pd
import pandas as pd
```

```python
datasetpath='/content/retail_transactions.csv'
```

```python
df=pd.read_csv(datasetpath)
```

```python
df
```

```python
from datetime import datetime
```

```python
df['TransactionDate']=pd.to_datetime(df['TransactionDate'])
```

```python
df.describe()
```

```python
#check for missing values in each column
missing_value=df.isnull().sum()
```

```python
missing_value
```

```python
missing_value.value_counts()
```

```python
#Ensure correct numerical formats for Quantity, Price, and Discount

df['Quantity']=pd.to_numeric(df['Quantity'])
df['Price']=pd.to_numeric(df['Price'])

df['Discount']=pd.to_numeric(df['Discount'])
df['Discount']=df['Discount'].fillna(0)
```

```python
df
```

```python
#Calculate NetPrice
df['NetPrice']=df['Price']*(1-df['Discount']/100)
#Calculate TotalSale
df['TotalSale']=df['Quantity']*df['NetPrice']
```

```python
df
```

```python
#Data Filtering
filtered_df=df[df['TotalSale']>=20]
```

```python
filtered_df
```

```python
#Data Type Conversion
filtered_df['PaymentType']=df['PaymentType'].astype('category')
```

```python
filtered_df['PaymentType']
```

```python
#Grouping and Aggregation
grouped_df = df.groupby('ProductCategory')[['Quantity', 'TotalSale']].sum().reset_index()
grouped_df = grouped_df.sort_values(by='TotalSale', ascending=False)
```

```python
top_3_categories=grouped_df.head(3)
```

```python
top_3_categories
```

```python
grouped_df=df.groupby(['StoreID','ProductCategory'])['TotalSale'].mean().reset_index()
```

```python
#Group the filtered data by both StoreID and ProductCategory

multi_grouped = filtered_df.groupby(['StoreID', 'ProductCategory']).agg(
    AverageTotalSale=('TotalSale', 'mean')
).reset_index()
```

```python
grouped_df
```

```python
multi_grouped
```

```python
# For each store(StoreID),determine which product category has the highest average TotalSale
store_insights = multi_grouped.loc[multi_grouped.groupby('StoreID')
['AverageTotalSale'].idxmax()]
```

```python
#prepare and print a summary DataFrame
summary_df = store_insights[['StoreID', 'ProductCategory', 'AverageTotalSale']]
```

```python
summary_df
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(filtered_df['Quantity'], filtered_df['NetPrice'], color='coral', alpha=0.5)
plt.xlabel('Quantity')
plt.ylabel('Net Price')
plt.title('Quantity vs Net Price')
plt.show()
```

```python
plt.bar(grouped_df.index, grouped_df['TotalSale'], color='skyblue')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.title('Total Sales per Product Category')
plt.show()
```
